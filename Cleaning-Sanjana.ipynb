```python
qog=pd.read_csv(link)
```

```python
#First, there are a bunch of columns we don't really need, so we will aim to create a new dataframe with only those we need
#Focus of data is the following variables
#bci_bci- The Bayesian Corruption Indicator
#egov_egov- E-Government Index
#cspf_sfi- State Fragility Index
```

```python
qognew=qog.filter(['cname','bci_bci','egov_egov','cspf_sfi'], axis=1)
```

```python
qognew
```

```python
#Now I have only the data I want but I should check for issues in the data
NumericColNames=qognew.iloc[:,1:3].columns
```

```python
badSymbols=[]
for columnName in NumericColNames:
    for cell in qognew[columnName]:
        try:
            float(cell)
        except:
            if cell not in badSymbols:
                badSymbols.append(cell)
```

```python
#Doesn't look like the numeric columns have any issues.
#Given the three variables I am looking at to understand quality of government, there are no categoricals.
qognew.dtypes
```

Merging With Other Files

```python
import pandas as pd

from urllib.request import urlopen

IEF=pd.read_pickle(urlopen("https://github.com/andresbas/DataScience/raw/master/IEF_Clean.pkl"),compression=None)

Env=pd.read_pickle(urlopen("https://github.com/Raslank/Pubpol-542/raw/master/environment.pkl"),compression=None)

QOG=pd.read_pickle(urlopen("https://github.com/sanjp0594/PUBPOL542-Data/raw/master/QOG_Cleaned.pkl"),compression=None)

GEI=pd.read_pickle(urlopen("https://github.com/shp9026/govanalytics/raw/master/gei.pkl"),compression=None)
```

```python
FirstMerge=Env.merge(IEF,left_on="Country",right_on="Country")
```

```python
FirstMerge
```

```python
SecondMerge=FirstMerge.merge(QOG,left_on="Country",right_on="cname")
```

```python
SecondMerge
```

```python
FinalMerge=SecondMerge.merge(GEI,left_on="Country_x",right_on="Country", how='outer')
```

```python
FinalMerge
```

```python
import rpy2

from rpy2.robjects import pandas2ri
pandas2ri.activate()

from rpy2.robjects.packages import importr

base = importr('base')
base.saveRDS(FinalMerge,file="TEAMdata.RDS")
```

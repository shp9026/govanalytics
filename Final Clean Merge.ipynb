```python
merger1=andres.merge(kamal,left_on="Country",right_on="Country",)#how='outer',indicator=True) 
```

```python
merger2=merger1.merge(sanjana,left_on="Country",right_on="cname",)#how='outer') 
```

```python
merger2.drop(columns='cname',inplace=True) 
```

```python
from rpy2.robjects import pandas2ri
pandas2ri.activate()

from rpy2.robjects.packages import importr

base = importr('base')
base.saveRDS(merger2,file="FinalMerge.RDS")
```

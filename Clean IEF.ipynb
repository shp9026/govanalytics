```python
IEF.GDPperCapita=IEF.GDPperCapita.str.replace("$","")
```

```python
IEF.GDPperCapita=IEF.GDPperCapita.str.replace("\s(\D+)","")
```

```python
badIEFSymbols=[] # list for bad symbols

NumericColNames=IEF.iloc[:,1:].columns # save names of columns with numeric data

for columnName in NumericColNames:# visit every column name
    for cell in IEF[columnName]:# visit every cell for that column
        try:
            float(cell) # try this
        except: # if not possible:            
            if cell not in badIEFSymbols:# if cell is not in the list                
                badIEFSymbols.append(cell)# add it to the list

# you get:
badIEFSymbols
```
